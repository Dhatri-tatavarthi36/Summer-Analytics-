### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Drop rows with missing horsepower values
df = df.dropna(subset=["horsepower"])

# Calculate the median mpg
median_mpg = df["mpg"].median()

# Filter cars with above-median mpg
above_median_mpg = df[df["mpg"] > median_mpg].copy()

# Calculate horsepower-to-weight ratio
above_median_mpg["hp_to_weight"] = above_median_mpg["horsepower"] / above_median_mpg["weight"]

# Find the maximum ratio
max_ratio = above_median_mpg["hp_to_weight"].max()

# Get the car(s) with the maximum ratio
best_cars = above_median_mpg[above_median_mpg["hp_to_weight"] == max_ratio]

print("Car(s) with best horsepower-to-weight ratio among above-median mpg cars:")
print(best_cars)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import seaborn as sns
# Drop rows with missing values in key columns
df = df.dropna(subset=["mpg", "model_year", "origin"])

# Group by model year and origin, then compute average mpg
grouped = df.groupby(["model_year", "origin"])["mpg"].mean().reset_index()

# Set Seaborn style
sns.set(style="whitegrid")

# Create the multi-line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=grouped, x="model_year", y="mpg", hue="origin", marker="o")

# Title and labels
plt.title("Average MPG Over the Years by Origin", fontsize=16)
plt.xlabel("Model Year", fontsize=12)
plt.ylabel("Average MPG", fontsize=12)
plt.legend(title="Origin")
plt.tight_layout()

plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
# Drop rows with missing horsepower or mpg values
df = df.dropna(subset=["horsepower", "weight", "mpg", "origin"])

# Set Seaborn style
sns.set(style="whitegrid")

# Create the scatterplot
plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(
    data=df,
    x="horsepower",
    y="weight",
    hue="origin",
    size="mpg",
    hue_order=["japan", "europe", "usa"],
    palette="deep",
    sizes=(20, 200),
    alpha=0.7,
    edgecolor="gray"
)

# Titles and labels
plt.title("Car Horsepower vs. Weight (Sized by MPG, Colored by Origin)", fontsize=16)
plt.xlabel("Horsepower", fontsize=12)
plt.ylabel("Weight", fontsize=12)
plt.legend(title="Origin", loc='upper right', bbox_to_anchor=(1.15, 1))
plt.tight_layout()

# Show the plot
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# Group by car name and count number of unique model years
multi_year_models = df.groupby("name")["model_year"].nunique()
multi_year_models = multi_year_models[multi_year_models > 1].index

# Filter to only those models
filtered = df[df["name"].isin(multi_year_models)]

# Group again to compute std and mean mpg, and count appearances
stats = filtered.groupby("name").agg(
    mpg_std=("mpg", "std"),
    mpg_mean=("mpg", "mean"),
    appearances=("mpg", "count")
).reset_index()

# Filter for standard deviation < 1.0
consistent_cars = stats[stats["mpg_std"] < 1.0]

# Sort by appearances (descending), then average mpg (descending)
consistent_cars_sorted = consistent_cars.sort_values(
    by=["appearances", "mpg_mean"], ascending=[False, False]
)
print("Consistent car models with low mpg variation:\n")
print(consistent_cars_sorted[["name", "appearances", "mpg_mean"]])
```
